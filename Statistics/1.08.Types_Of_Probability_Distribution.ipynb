# 1. Normal Distribution - Continuous

A normal distribution, also known as a Gaussian distribution, is a probability distribution that is symmetric around its mean. It is characterized by a bell-shaped curve where most of the data points are clustered around the mean, with progressively fewer data points as you move away from the mean in either direction. The normal distribution is parametrized by two parameters: the mean (μ) and the standard deviation (σ), which determine the center and spread of the distribution, respectively.
- Youtube Link: https://www.youtube.com/watch?v=mtbJbDwqWLE


```python
# Parameter for the Normal Distributon
mu = 0  # Parameter Mean
sigma = 1  # Parameter Standard Deviation

# Generate random data from a normal distribution
data = np.random.normal(mu, sigma, size=100)
# print(data)
df = pd.DataFrame({"Values": data})

# Generate data points for the x-axis
x = np.linspace(start= mu-3*sigma, stop=mu+3*sigma, num=100)

# Calculate the Probability Density Function for Each Data Points
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Create a plot of the normal distribution
plt.figure(figsize=(12, 5))
plt.hist(df['Values'], bins=20, density=True, alpha=0.75)
plt.plot(x, pdf, label="Normal Distributon Plot", color="red")
plt.title("Normal Distributon")
plt.xlabel("X-Axis")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# @CDF- Cumulative Density Function
# Value for which you want to find the cumulative probability
z = 0.15
cdf = stats1.norm.cdf(z, loc=0, scale=1)
print(f"CDF at z = {z} is: {cdf}")
```

### Properties of Normal Distribution:
The normal distribution, also known as the Gaussian distribution, is one of the most important probability distributions in statistics. It has several key properties:

1. **Symmetry**: The normal distribution is symmetric around its mean (μ). This means that the probability density function (PDF) of the distribution is the same on both sides of the mean, and the curve is bell-shaped.

2. **Bell-shaped Curve**: The probability density function forms a bell-shaped curve, which is often referred to as a "normal curve" or a "bell curve." It is unimodal, meaning it has a single peak at the mean.

3. **Mean and Median**: The mean (μ) and median are equal in a normal distribution, and they both coincide at the center of the distribution. This is a result of the symmetry of the distribution.

4. **Standard Deviation**: The standard deviation (σ) determines the spread or dispersion of the distribution. Larger values of σ result in wider and flatter curves, while smaller values of σ result in narrower and taller curves.

5. **68-95-99.7 Rule**: This empirical rule states that approximately 68% of the data falls within one standard deviation of the mean (μ ± σ), approximately 95% falls within two standard deviations (μ ± 2σ), and about 99.7% falls within three standard deviations (μ ± 3σ).

6. **Empirical Rule**: The normal distribution follows the empirical rule, which makes it a valuable distribution in practice for modeling various natural phenomena and making statistical inferences.

7. **Infinitely Extensible Tails**: The tails of the normal distribution extend infinitely in both directions, meaning that there is a very low but non-zero probability of observing values that are extremely far from the mean.

8. **Parameterization**: The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ). These parameters dictate the location and scale of the distribution, respectively.

9. **Additivity**: When two independent random variables are normally distributed, their sum (or linear combination) is also normally distributed. This property makes the normal distribution particularly useful in many mathematical and statistical applications.

10. **Central Limit Theorem**: The normal distribution plays a central role in the Central Limit Theorem, which states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution of those variables.

Due to these properties, the normal distribution is widely used in various fields, including statistics, finance, natural sciences, social sciences, and engineering, as it serves as a good approximation for many real-world phenomena.


CDF: The Cumulative Distribution Function (CDF) of a normal distribution provides the probability that a random variable following that distribution will be less than or equal to a specific value. The CDF of the normal distribution is denoted by Φ(z), where "z" represents the value for which you want to find the cumulative probability.



### Z Socore or Standard Normal Variate

- Youtube Video Link: https://www.youtube.com/watch?v=2tuBREK_mgE
A Standard Normal Variate, often referred to simply as a "standardized random variable" or "z-score," is a transformed value obtained by subtracting the mean (μ) of a random variable and dividing by its standard deviation (σ). Standardization is a common statistical technique used to compare or analyze data on a common scale, particularly when dealing with different distributions or units of measurement. 


```python
# Define your dataset as a NumPy array
data = np.array([10, 15, 20, 25, 30])

# Caclculate Mean and Standard variation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate z-scores for each data point
z_scores = (data - mean) / std_dev

print("Original Data:", data)
print("Z-Scores:", z_scores)
```

# 2. Unifrom Distributon - Continuous

Youtube Link: https://www.youtube.com/watch?v=UC-CBUSQXAo

A uniform distribution, also known as a rectangular distribution, is a probability distribution in which all values within a specified range are equally likely to occur. In other words, every value in the range has the same probability of being selected. The probability density function (PDF) of a uniform distribution is constant within the range and zero outside the range.

```python
# Step 1: Generate random data following a uniform distribution
np.random.seed(42)  # Set a seed for reproducibility
low = 0
high = 1
num_samples = 1000
uniform_data = np.random.uniform(low, high, num_samples)

df = pd.DataFrame({"Values": uniform_data})

# Step 2: Create a histogram to visualize the uniform distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Values'], bins=20, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.grid(True)

# Plot the theoretical probability density function (PDF)
x = np.linspace(0, 1, 100)  # Generate x-values for the PDF line
pdf = np.ones_like(x)  # Uniform PDF is constant
plt.plot(x, pdf, 'r-', lw=2, label='PDF (Theoretical)')

plt.legend()
plt.show()
```

# 3. Bernauli Distribution - Discrete

- Youtube Link: https://www.youtube.com/watch?v=nl9WiZMZnYs
- The Bernoulli distribution is a discrete probability distribution that models a random experiment with two possible outcomes: success and failure. It is characterized by a single parameter, often denoted as "p," which represents the probability of success. The probability of failure is given by 1 - p.

```python
# Parameters
p = 0.3  # Probability of success

# Generate Bernoulli random variables (0 or 1)
num_samples = 1000
bernouli_data = np.random.choice([0, 1], size=num_samples, p=[1 - p, p])
bernouli_data

# Plot the histogram
plt.hist(bernouli_data, bins=[-1.5, 0.5, 1.5], density=True, alpha=0.75, edgecolor="black")
plt.xticks([0,1], ["Fialure (0)", "Success (1)"])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution (p={p})')

plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Binomial Distribution - Discrete

- Youtube Link: https://www.youtube.com/watch?v=gT26Y_VJmOM
- The binomial distribution is a discrete probability distribution that models the number of successes (usually denoted as "k") in a fixed number of independent Bernoulli trials (experiments with two possible outcomes: success and failure). Each trial is characterized by a probability of success denoted as "p." The binomial distribution is described by two parameters: "n" (the number of trials) and "p" (the probability of success in each trial).

```python
# Parameters for the binomial distribution
n = 10 # The number of trails
p = 0.5 # Probability of success on a single trial

# Generate a sample binomial distribution
x = np.arange(0, n+1) # Possible number of successes (0 to n)
pmf_values = binom.pmf(x, n, p)  # Probability mass function values
pmf_values

# Plot the probability mass function
plt.figure(figsize=(12, 6))
plt.bar(x, pmf_values, label=f"Binomial PMF (n={n}, p={p})")
plt.xlabel("Number of Success")
plt.ylabel("Probability")
plt.title("Binomial Distribution PMF")

plt.grid(True)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

# 5. Poisson Distribution - Discrete

- Youtube Link: https://www.youtube.com/watch?v=BbLfV0wOeyc
- The Poisson distribution is a discrete probability distribution that models the number of events or occurrences that happen within a fixed interval of time or space, assuming that these events occur at a constant average rate and are independent of the time since the last event. It is named after the French mathematician Siméon Denis Poisson.

Key characteristics and properties of the Poisson distribution:

1. **Parameter**: The Poisson distribution is parameterized by a single positive real number denoted as \(\lambda\) (lambda), which represents the average rate at which events occur in the given interval. It can also be interpreted as the expected number of events in the interval.


```python
np.random.seed(42)

# Set the average rate (lambda) for the Poisson distribution
lambda_value = 3

# Generate a sample of Poisson-distributed data
sample_size = 1000
poisson_data = np.random.poisson(lambda_value, size=sample_size)
poisson_data

# Create a Pandas DataFrame
df = pd.DataFrame({"Poisson_data": poisson_data})

# Plot a histogram of the Poisson data
plt.hist(df['Poisson_data'], bins=20, density=True, alpha=0.75, edgecolor="black")
plt.title('Poisson Distribution')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')

# Plot the PDF line for the Poisson distribution
x = np.arange(0, 15)
pmf_poisson = poisson.pmf(x, lambda_value) # Compute PMF for Poisson distribution
plt.plot(x, pmf_poisson, color="red", marker="o", linestyle="-", lw=2)

plt.grid()
plt.show()
```

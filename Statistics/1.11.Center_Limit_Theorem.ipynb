# Central Limit Theorem

The Central Limit Theorem (CLT) is a fundamental concept in probability theory and statistics. It states that when you repeatedly draw random samples of a large enough size from any population, regardless of its underlying distribution, the distribution of the sample means will approximate a normal (Gaussian) distribution. In simpler terms, the CLT tells us that if we take many samples from a population and calculate the mean of each sample, those sample means will tend to follow a normal distribution.

Here are some key points about the Central Limit Theorem:

Sample Size: The Central Limit Theorem works best when the sample size is sufficiently large. While there isn't a strict rule for what constitutes "sufficiently large," a common guideline is that a sample size of at least 30 is often adequate.

Independence: The samples must be drawn independently. Each observation in a sample should not be dependent on or influenced by the others.

Population Distribution: The original population can have any distribution, not necessarily a normal distribution. It can be uniform, exponential, bimodal, or any other distribution.



Problem:

Suppose you are a quality control manager at a chocolate factory. You are responsible for ensuring that the weight of chocolate bars produced follows the desired specifications. The company claims that the mean weight of a chocolate bar is 50 grams with a standard deviation of 3 grams.

To maintain quality control, you decide to take a random sample of 36 chocolate bars from the production line and calculate their average weight. You want to use the Central Limit Theorem to make inferences about the distribution of the sample means.

a) According to the Central Limit Theorem, what can you expect the distribution of the sample means to be? Provide the mean and standard deviation of this distribution.

b) If you find that the sample mean weight of the 36 chocolate bars is 51.5 grams, what is the probability that the true population mean weight is still 50 grams? Calculate this using the z-score and standard normal distribution.

c) What sample size would you need to be 95% confident that your sample mean weight estimate is within 0.5 grams of the true population mean weight of chocolate bars?

(Note: You can assume that the weights of chocolate bars in the population are normally distributed.)

```python
# Solve this Problem
# Population Parameter
population_mean = 50
population_std = 3
sample_size = 36

# a) Generate a sample of 36 chocolate bar weights and calculate the sample mean
np.random.seed(42)
sample = np.random.normal(population_mean, population_std, sample_size)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1) #ddof=1 for sample standard deviation
print("a) Sample Mean Weights: ", sample_mean, "grams")

# b) Calculate the z-score and find the probability
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
probability = 1 - norm.cdf(z_score)
print("b) Probability that the true population mean is still 50 grams:", probability)

# c) Calculate the required sample size for a 95% confidence interval with a margin of error of 0.5 grams
confidence_level= 0.95
alpha = 1 - confidence_level
margin_of_error = 0.05

# Calculate the critical z-value for the desired confidence level
critical_z = norm.ppf(1 - alpha/2 )

# Calculate the required sample size using the formula: n = (critical_z * (population_std / margin_of_error)) ^ 2
required_sample_size = ((critical_z * (population_std / margin_of_error)) **2).astype(int)
print("c) Required Sample Size for a 95% confidence interval with a margin of error of 0.5 grams:", required_sample_size)

print("")
# Plot a histogram of the sample
plt.hist(sample, bins = 15, density=True, alpha=0.75, edgecolor="black", label="Sample Data")
plt.xlabel('Chocolate Bar Weight (grams)')
plt.ylabel('Probability Density')
plt.title('Distribution of Chocolate Bar Weights in the Sample')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

A sampling distribution is a probability distribution that shows the possible values of a statistic (such as the mean, variance, standard deviation, or any other measure) obtained from different samples of the same size taken from the same population. In simpler terms, it represents how a specific statistic varies when calculated from various random samples drawn from a population.

Here are some common types of sampling methods:

1. Simple Random Sampling: In simple random sampling, each member of the population has an equal chance of being selected. This is typically done using random number generators or drawing lots. It is considered one of the most unbiased and straightforward sampling methods.

2. Stratified Sampling: Stratified sampling involves dividing the population into subgroups or strata based on a specific characteristic (e.g., age, gender, income level), and then random samples are taken from each stratum. This ensures that each subgroup is represented in the sample, making it useful when you want to analyze subpopulations separately.

3. Systematic Sampling: Systematic sampling involves selecting every nth member from a list or sequence of the population. The first member is randomly chosen, and then subsequent members are selected at regular intervals. This method is simple and efficient when a complete list of the population is available.

4. Cluster Sampling: In cluster sampling, the population is divided into clusters or groups, and a random sample of clusters is selected. Then, all members within the selected clusters are included in the sample. Cluster sampling is often used when it's difficult or costly to obtain a complete list of the population.

Problem:

You are conducting a survey to estimate the average monthly household electricity consumption in a particular city. The city has a population of 10,000 households. Due to budget constraints, you can only survey a sample of 100 households.

4. **Confidence Interval**: Calculate a 95% confidence interval for the true population mean monthly electricity consumption based on your sample data. Interpret the interval.

5. **Sample Size Consideration**: Suppose you want to reduce the margin of error in your estimate of the population mean to 25 kWh. Calculate the minimum sample size required to achieve this level of precision while maintaining a 95% confidence level.

6. **Bias Assessment**: Discuss any potential sources of bias in your survey and how they might affect the accuracy of your estimate of the population mean electricity consumption.

(Note: You can assume that the monthly electricity consumption follows an approximately normal distribution in the population.)

```python
# Solve Problem
# Given data
population_size = 10000
sample_size = 100
sample_mean = 450  # kWh
sample_std = 75  # kWh
confidence_level = 0.95
margin_of_error = None  # We'll calculate this later

# 1. Sampling Method (Simple Random Sampling)
# Generate a random sample of household IDs from the population
np.random.seed(42)  # for reproducibility
sample_household_ids = np.random.choice(population_size, sample_size, replace=False)
print(sample_household_ids)
# 2. Sample Collection

# 3. Sampling Distribution of the Sample Mean
# Calculate the standard error of the sample mean
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the mean and standard deviation of the sampling distribution
sampling_mean = sample_mean
sampling_std = standard_error

# 4. Confidence Interval
# Calculate the margin of error for the given confidence level
z_critical = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_critical * (sampling_std)

# Calculate the confidence interval
confidence_interval = (sampling_mean - margin_of_error, sampling_mean + margin_of_error)

# 5. Sample Size Consideration (Calculate minimum sample size for desired margin of error)
desired_margin_of_error = 25  # kWh

# Rearrange the formula for margin of error to solve for sample size
sample_size_required = ((z_critical * sampling_std) / desired_margin_of_error) ** 2

# Display the results
print("1. Sampling Method: Simple Random Sampling")
print("2. Sample Collection: Sample Mean =", sample_mean, "kWh, Sample Std Dev =", sample_std, "kWh")
print("3. Sampling Distribution of Sample Mean:")
print("   - Sampling Mean =", sampling_mean, "kWh")
print("   - Sampling Std Dev =", sampling_std, "kWh")
print("4. Confidence Interval (95%):", confidence_interval)
print("5. Minimum Sample Size Required for 25 kWh Margin of Error:", int(np.ceil(sample_size_required)))
print("")

# Plot the sampling distribution
x = np.linspace(sampling_mean - 4 * sampling_std, sampling_mean + 4 * sampling_std, 1000)
y = stats.norm.pdf(x, sampling_mean, sampling_std)
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Sampling Distribution")
plt.axvline(confidence_interval[0], color='r', linestyle='--', label="95% Confidence Interval")
plt.axvline(confidence_interval[1], color='r', linestyle='--')
plt.xlabel("Sample Mean (kWh)")
plt.ylabel("Probability Density")
plt.title("Sampling Distribution of Sample Mean")
plt.legend()
plt.grid(True)
plt.show()
```

# Standard Error

Youtube Link: https://www.youtube.com/watch?v=UE4iLGXGlI0

Standard error is a statistical measure that quantifies the amount of variability or uncertainty associated with the sample mean when estimating a population parameter. In other words, it tells you how much the sample mean is likely to vary from the true population mean.

```python
# Calculate Standard Error
data = np.array([23, 45, 56, 34, 67, 12, 78, 54, 32, 67])

# Calculate the standard error
std_deviation = np.std(data)
sample_size = len(data)
standard_error = std_deviation / np.sqrt(sample_size)

print("Standard Error is: ", standard_error)
```

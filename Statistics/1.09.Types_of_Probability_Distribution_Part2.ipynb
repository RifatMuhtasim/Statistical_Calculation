While the binomial distribution describes the number of successes (usually denoted as "k") in a fixed number of independent Bernoulli trials (experiments with two possible outcomes, usually success and failure), the multinomial distribution deals with the number of occurrences of each of multiple discrete outcomes in a fixed number of trials.

Key characteristics of a multinomial distribution:

1. Multiple Outcomes: In a multinomial experiment, there are more than two possible outcomes, often denoted as categories or classes. Each outcome has a probability of occurring.

2. Fixed Number of Trials: The experiment consists of a fixed number of trials or events, denoted as "n." Each trial results in one of the possible outcomes.

3. Independent Trials: Each trial is independent of the others. The outcome of one trial does not affect the outcomes of other trials.

4. Probability Vector: The probabilities of each outcome are represented as a probability vector. This vector specifies the probability of each outcome occurring in a single trial.

A log-normal distribution is a continuous probability distribution of a random variable where the natural logarithm of that variable follows a normal distribution. In simpler terms, if you take the logarithm of a random variable that follows a log-normal distribution, the resulting values will be normally distributed.

- x is the random variable that follows a log-normal distribution.
- μ is the mean of the natural logarithm of x.
- σ is the standard deviation of the natural logarithm of x.
- e is the base of the natural logarithm (approximately equal to 2.71828).

Some key properties and characteristics of the log-normal distribution include:

1. Positivity: Values of x are strictly positive since the natural logarithm of a non-zero number is always finite.

2. Skewness: The log-normal distribution is positively skewed, meaning it has a long right tail. This is because the logarithm of a variable can never be negative, resulting in more values concentrated on the lower end.

3. Moments: The moments of the log-normal distribution (e.g., mean, variance) can be expressed in terms of μ and σ. The mean and variance of the log-normal distribution in terms of the original variable x are not straightforward to calculate and depend on the parameters μ and σ.

4. Applications: The log-normal distribution is often used to model phenomena where the values are inherently positive and vary multiplicatively rather than additively. It is commonly applied in fields such as finance (e.g., modeling stock prices), biology (e.g., modeling the size of biological organisms), and geology (e.g., modeling the distribution of mineral deposits).

5. Transformation: When dealing with data that follows a log-normal distribution, it is sometimes beneficial to transform the data by taking the natural logarithm of the observations. This can make the data more amenable to analysis and visualization, as it transforms the distribution into a normal (Gaussian) distribution.

It's important to note that while the log-normal distribution is useful in certain contexts, it may not be appropriate for all types of data. It should be chosen based on a theoretical understanding of the underlying process being modeled.

# 10. Pareto Distribution - Continuos

Youtube Link: https://www.youtube.com/watch?v=SF2dLUB6t28

The Pareto distribution is a probability distribution used to model phenomena where a small number of items account for the majority of the occurrences or values. It is named after the Italian economist Vilfredo Pareto, who observed that a small fraction of the population owned a large proportion of the wealth, which led to the development of this distribution. The Pareto distribution is characterized by a heavy tail, meaning that extreme values are more likely to occur compared to many other distributions.

Here's the formal definition of the Pareto distribution:

**Probability Density Function (PDF):**

The probability density function of the Pareto distribution is defined as:

```
f(x; α, xₘ) = α * xₘ^α / x^(α+1)
```

Where:
- `x` is the random variable.
- `α` is the shape parameter, often called the Pareto index. It must be greater than 1.
- `xₘ` is the minimum possible value of `x`, also known as the scale parameter.

Key points about the Pareto distribution:

1. **Heavy Tail:** The Pareto distribution is known for its heavy tail, which means that it has a slowly decreasing tail, and extreme values have a higher probability of occurring.

2. **Power Law Distribution:** It is a specific case of a power-law distribution, where the probability of an event is inversely proportional to a power of its size.

3. **Applications:** The Pareto distribution is used in various fields, including economics (where it's often used to describe income and wealth distributions), finance, natural sciences, and more.

4. **Mean and Variance:** The mean (expected value) and variance of a Pareto-distributed random variable depend on the shape parameter `α`. For `α > 1`, the mean is finite, but the variance may be infinite.

5. **Truncated Pareto:** In some cases, a truncated Pareto distribution is used, which restricts the range of possible values to a finite interval.

The Pareto distribution is a valuable tool for modeling situations where a small fraction of entities or events accounts for a significant portion of the total, making it particularly useful in understanding and analyzing phenomena characterized by extreme inequality or concentration.

````python
# Define the parameters of the Pareto distribution
alpha = 2.0 # Shape parameter (must be greater than 1)
xmin = 1.0 # # Minimum possible value
sample_size = 100

# Generate Random Sample from Pareto Distributin
samples = np.random.pareto(alpha, sample_size) + xmin
samples

df = pd.DataFrame({"data": samples})
df

# Show the Histogram
plt.hist(df['data'], bins=30, density=True, alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Pareto Distribution')

# Overlay a line plot of the PDF
x = np.linspace(min(samples), max(samples), 100)
pdf = (alpha * xmin**alpha) / x**(alpha+1)
print(pdf)
plt.plot(x, pdf, 'r-', linewidth=2, label="PDF")

plt.grid(True)
plt.show()
````

The Chi-Square distribution is a probability distribution that arises in statistics and probability theory. It is closely related to the Gaussian (normal) distribution and is used in various statistical tests and estimations, particularly in the context of hypothesis testing and confidence intervals.

Key characteristics and properties of the Chi-Square distribution include:

1. **Degrees of Freedom (df):** The Chi-Square distribution is characterized by a parameter called degrees of freedom (df), denoted as "k." The degrees of freedom determine the shape and scale of the distribution. The Chi-Square distribution with k degrees of freedom is often denoted as χ²(k).

2. **Shape:** The shape of the Chi-Square distribution depends on the degrees of freedom. As k increases, the distribution becomes less skewed and approaches a bell-shaped curve (similar to the normal distribution).

3. **Non-Negativity:** The values of a random variable following a Chi-Square distribution are always non-negative.

4. **Skewness:** The Chi-Square distribution is positively skewed, particularly for lower values of k.

5. **Applications:** The Chi-Square distribution is commonly used in statistical tests, such as the Chi-Square test for independence (used to assess the relationship between categorical variables), the Chi-Square goodness-of-fit test (used to compare observed and expected frequencies), and in the construction of confidence intervals for population variances.

6. **Probability Density Function (PDF):** The probability density function of the Chi-Square distribution is given by:

   ```
   \[f(x; k) = (1 / (2^(k/2) * Γ(k/2))) * x^(k/2 - 1) * e^(-x/2)\]
   ```

   Where:
   - \(x\) is the random variable.
   - \(k\) is the degrees of freedom.
   - \(Γ\) is the gamma function.

8. **Mean and Variance:** The mean (expected value) of the Chi-Square distribution with k degrees of freedom is \(k\), and the variance is \(2k\).

The Chi-Square distribution is a fundamental concept in statistical analysis, and its properties are extensively used in hypothesis testing and statistical inference. It plays a crucial role in determining critical values and p-values in statistical tests, which are essential tools in making decisions based on data analysis.

Student's t-distribution, often referred to simply as the t-distribution, is a probability distribution that plays a crucial role in statistical inference, particularly in situations involving small sample sizes when the population standard deviation is unknown. The t-distribution was introduced by William Sealy Gosset, who published under the pseudonym "Student."

Key characteristics and properties of the t-distribution include:

1. **Shape:** The t-distribution is symmetric and bell-shaped, similar to the normal distribution. However, its shape depends on a parameter called degrees of freedom (df).

2. **Degrees of Freedom (df):** The degrees of freedom determine the shape and characteristics of the t-distribution. The greater the degrees of freedom, the closer the t-distribution resembles a normal distribution.

3. **Heavier Tails:** Compared to the normal distribution, the t-distribution has heavier tails. This means that the t-distribution has more probability in its tails, making it more robust when dealing with outliers or when the sample size is small.

4. **Applications:** The t-distribution is widely used in hypothesis testing, confidence interval estimation, and statistical modeling when working with small sample sizes. It is particularly important when estimating population parameters, such as the population mean, with limited data.

5. **Probability Density Function (PDF):** The probability density function of the t-distribution with df degrees of freedom is given by:

   \[f(t; df) = \frac{\Gamma\left(\frac{df + 1}{2}\right)}{\sqrt{df\pi}\Gamma\left(\frac{df}{2}\right)} \left(1 + \frac{t^2}{df}\right)^{-\frac{df + 1}{2}}\]

   Where:
   - \(t\) is the random variable following the t-distribution.
   - \(df\) is the degrees of freedom.
   - \(\Gamma\) represents the gamma function.

6. **Mean and Variance:** The mean of the t-distribution is 0, and its variance is \(\frac{df}{df - 2}\) for \(df > 2\). For \(df = 1\) or \(df = 2\), the variance is undefined.

The t-distribution is especially valuable when working with sample data from a population with an unknown population standard deviation. It allows statisticians to make inferences about population parameters and construct confidence intervals that take into account the variability introduced by small sample sizes.

````python
# Define degrees of freedom
degrees_of_freedom = 10 

# Generate random samples from the t-distribution
sample_size = 1000
t_samples = np.random.standard_t(degrees_of_freedom, size= sample_size)
t_samples

df = pd.DataFrame({"T_Samples": t_samples})

# Create a histogram to visualize the t-distribution
plt.hist(df['T_Samples'], bins = 30, density=True, alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f"Student's t-Distribution (df={degrees_of_freedom})")

# Overlay a line plot of the PDF
x_values = np.linspace(-5, 5, 1000)
pdf_values = t.pdf(x_values, degrees_of_freedom)
plt.plot(x_values, pdf_values, 'r-', lw=2, label="PDF")

plt.legend()
plt.tight_layout()
plt.show()
````

# Overview of Probability Distribution


1. **Normal Distribution**:
   - Also known as the bell curve.
   - Symmetrical and characterized by its mean and standard deviation.
   - Many natural phenomena follow this distribution.

2. **Uniform Distribution** (Continuous):
   - All values within a specified range are equally likely.
   - Constant probability density within the interval and zero outside.

3. **Bernoulli Distribution**:
   - Models binary outcomes, such as success (1) or failure (0) in a single trial.

4. **Binomial Distribution**:
   - Represents the number of successes in a fixed number of independent Bernoulli trials.

5. **Poisson Distribution**:
   - Models the number of events occurring in a fixed interval when events are rare and independent.

6. **Beta Distribution**:
   - Constrained to the interval [0, 1] and often used to model probabilities and proportions.

7. **Gamma Distribution**:
   - Generalizes the exponential distribution and can model the sum of multiple exponential random variables.

8. **Multinomial Distribution**:
   - Generalizes the binomial distribution to more than two categories.
   - Describes the probability distribution of categorical outcomes with more than two possibilities.

9. **Log-Normal Distribution**:
   - The natural logarithm of the random variable follows a normal distribution.
   - Used to model variables that are naturally positive and have a wide range of values.
     
10. **Pareto Distribution**:
    - The Pareto distribution is a continuous probability distribution characterized by a heavy tail. It's often used to model phenomena         where a small number of observations have a significant impact, such as wealth distribution, income distribution, and the popularity of items on the internet.
    
12. **Chi-Square Distribution**:
    - Used in hypothesis testing and confidence interval construction.
    - Often applied in the context of the chi-squared test.

13. **Student's T Distribution**:
    - Used in hypothesis testing when the sample size is small and the population standard deviation is unknown.
    - Heavier tails compared to the normal distribution, allowing for better handling of small sample uncertainty.

Each of these distributions serves different purposes in statistics and probability theory, and they find applications in various fields depending on the characteristics of the data being modeled.
